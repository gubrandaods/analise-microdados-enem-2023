## 3.0 Limpeza dos Dados

```python
# 1. Contagem absoluta de valores em branco por coluna
missing_counts = df_micro.isnull().sum()

# 2. Percentual de valores em branco por coluna
missing_perc = df_micro.isnull().mean() * 100

# 3. Combinar em um DataFrame organizado
missing_df = pd.concat(
    [missing_counts, missing_perc],
    axis=1,
    keys=['missing_count', 'missing_pct']
).sort_values(by='missing_count', ascending=False)
```
